```python
# Function to read and extract text from a PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text() + "\n"  # Add a newline character to separate pages
    return text
```

```python
# Path to your PDF document
pdf_path = "SAMPLE APPRAISAL.pdf"
# Extract text from the PDF
text = extract_text_from_pdf(pdf_path)
```

```python
import re

def is_header(line):
    # Check for all caps words or specific enumeration patterns
    return line.isupper() or re.match(r'\d+\) .+', line) or re.match(r'\d+\. .+', line)

def segment_text(text):
    lines = text.split('\n')
    sections = []
    current_section = []
    blank_line_count = 0  # Count consecutive blank lines

    for line in lines:
        if not line.strip():  # Check for blank lines
            blank_line_count += 1
        else:
            if is_header(line) or blank_line_count > 1:
                if current_section:
                    sections.append('\n'.join(current_section).strip())
                    current_section = []
            blank_line_count = 0  # Reset blank line count
            current_section.append(line)

    if current_section:  # Add the last section if any
        sections.append('\n'.join(current_section).strip())

    return sections

# Example usage
text = extract_text_from_pdf(pdf_path)
sections = segment_text(text)
for i, section in enumerate(sections, start=1):
    print(f"Section {i}:\n{section}\n")
```
